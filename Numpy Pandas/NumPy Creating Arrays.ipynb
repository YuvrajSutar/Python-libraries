# Numpy Arrays

NumPy arrays are the main way we will use Numpy throughout the course. Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-d arrays and matrices are 2-d (but you should note a matrix can still have only one row or one column).



The most important object defined in NumPy is an N-dimensional array type called ndarray. It describes the collection of items of the same type. 

Let's begin our introduction by exploring how to create NumPy arrays.

## Creating NumPy Arrays



```python
# one dimension 

my_array = np.array([1,2,3,4,5]) 
my_array.shape
```

```python
my_list = [1,2,3]
my_list
```

```python
np.array(my_list)
```

```python
# more than one dimensions 
my_matrix = [[1,2,3],[4,5,6],[7,8,9]]
my_matrix
```

```python
mat=np.array(my_matrix)
mat
```

## Built-in Methods

There are lots of built-in ways to generate Arrays

### arange

Return evenly spaced values within a given interval.



### Some Exapmles



```python
np.arange(2,10,3)
```

```python
np.arange(0,11,4)
```

### zeros and ones

Generate arrays of zeros or ones



```python
np.zeros(50)
```

```python
np.zeros((5,5))
```

```python
np.ones(3)
```

```python
np.ones((3,3))
```

### linspace
Return evenly spaced numbers over a specified interval.

In this function, instead of step size, the number of evenly spaced values between the interval is specified.



```python
np.linspace(1,100,10)
```

```python
np.linspace(0,10,5)
```

## Random 

Numpy also has lots of ways to create random number arrays:



### rand
Create an array of the given shape and populate it with
random samples from a uniform distribution
over ``[0, 1)``.

```python
np.random.rand(1,100,10)
```

```python
np.random.rand(5,5)
```

```python
np.random.randn(50)
```

### randn



Return a sample (or samples) from the "standard normal" distribution. Unlike rand which is uniform:

```python
np.random.randn(2)
```

```python
np.random.randn(5,5).shape
```

### randint
Return random integers from `low` (inclusive) to `high` (exclusive).

```python
np.random.randint(1,20)
```

```python
np.random.randint(1,101,20)
```
